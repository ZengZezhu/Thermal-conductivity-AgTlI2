```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kappadata = np.loadtxt('exp_cooling_T1')
temp = kappadata[:,0]; kappa = kappadata[:,1]; err = kappadata[:,2]
```

```python
k1 = np.loadtxt('Tl3VSe4')
k2 = np.loadtxt('AgI')
k3 = np.loadtxt('InTe')
k4 = np.loadtxt('AgCl')
k6 = np.loadtxt('PbTe')
k7 = np.loadtxt('PbSe')
k8 = np.loadtxt('AgCrSe2')
k9 = np.loadtxt('AgBr')
```

```python
plt.figure(figsize=(4.1, 4))
plt.xlabel('Temperature (K)',fontsize=13)
plt.ylabel(r'$\kappa_{\rm Exp}$ (Wm$^{-1}$K$^{-1}$)',fontsize=13,labelpad=-7)

plt.errorbar(temp, kappa, yerr=err*0.4, fmt='p', color='purple', marker='^', linestyle=':', label=r'$\rm AgTlI_2$'+' (This work-PPMS)', fillstyle='none', mec='#440154', ecolor='tab:gray', alpha=0.6)


plt.scatter(k1[:,0], k1[:,1], label=r'$\rm Tl_{3}VSe_{4}$', marker='D', color='#404387', facecolors='none', edgecolors='#404387')
plt.scatter(np.exp(k2[:,0]), np.exp(k2[:,1])+0., label=r'$\rm AgI$', marker='*', color='#29788E', facecolors='none', edgecolors='#29788E')
plt.scatter(k9[:,0], k9[:,1], label=r'$\rm AgBr$', marker='>', color='tab:blue', facecolors='none', edgecolors='tab:blue')
plt.scatter(k4[:,0], k4[:,1], label=r'$\rm AgCl$', marker='>', color='#44C463', facecolors='none', edgecolors='#44C463')
plt.scatter(k3[:,0], k3[:,1], label=r'$\rm InTe$', marker='d', color='#22A884', facecolors='none', edgecolors='#22A884')
plt.scatter(k7[:,0], 1./k7[:,1], label='PbSe', marker='o', color='#7EB138', facecolors='none', edgecolors='#7EB138')
plt.scatter(k6[:,0], k6[:,1], label=r'$\rm PbTe$', marker='<', color='#7EB138', facecolors='none', edgecolors='#FDE724')
plt.scatter(k8[:,0], k8[:,1], label=r'$\rm AgCrSe_{2}$', marker='v', color='tab:orange', facecolors='none', edgecolors='tab:orange')



plt.xlim(4,450)
plt.ylim(0.1,8)
plt.xticks([10,50,100,150,200,250,300,350],fontsize=13)
plt.yticks([0.1,1,2,3,4,5,6,7],fontsize=13)
plt.xscale('log') 
plt.yscale('log')
plt.legend(frameon=False)
plt.tight_layout()
#plt.savefig('expkappa.pdf')
```
